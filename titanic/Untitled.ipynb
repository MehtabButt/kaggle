```python
import pandas as pd
from sklearn.preprocessing import StandardScaler 
from sklearn.svm import SVC
```

```python
train_data = pd.read_csv("train.csv")
train_data.head()
```

```python
#droping unnecessary features
train_data.drop(["Name", "Ticket", "Cabin", "Embarked"], axis=1, inplace=True)
train_data.head()
```

```python
# checking missing values
train_data.isnull().sum()
```

```python
#filling migging feature values 
train_data.Age.fillna(train_data.Age.mean(), inplace=True)
train_data.isnull().sum()
```

```python
#converting string feature 'Sex' to integer
sex_dummies = pd.get_dummies(train_data["Sex"], drop_first=True)
train_data = pd.concat([train_data, sex_dummies], axis=1)
train_data.drop(['Sex'], axis=1, inplace=True)
train_data.head()
```

```python
std_scaler = StandardScaler()
features_scale = ['Age', 'Fare']
train_data[features_scale] = std_scaler.fit_transform(train_data[features_scale])
train_data.head()
```

```python
y = train_data.Survived
x = train_data.drop(['Survived'], axis=1)
```

```python
model_svc = SVC(C=100, kernel='rbf')
model_svc.fit(x, y)
```

```python
test_data = pd.read_csv("test.csv")
test_data.head()
```

```python
test_data.drop(["Name", "Ticket", "Cabin", "Embarked"], axis=1, inplace=True)
test_data.head()
```

```python
test_data.isnull().sum()
```

```python
test_data.Age.fillna(test_data.Age.mean(), inplace=True)
test_data.Fare.fillna(test_data.Fare.mean(), inplace=True)
train_data.isnull().sum()
```

```python
sex_dummies = pd.get_dummies(test_data["Sex"], drop_first=True)
test_data = pd.concat([test_data, sex_dummies], axis=1)
test_data.drop(['Sex'], axis=1, inplace=True)
test_data.head()
```

```python
test_data[features_scale] = std_scaler.fit_transform(test_data[features_scale])
test_data.head()
```

```python
y_pred = model_svc.predict(test_data)
```

```python
submission = pd.DataFrame({
    "PassengerId": test_data["PassengerId"],
    "Survived": y_pred
})
submission.to_csv("submission.csv", index=False)
```
